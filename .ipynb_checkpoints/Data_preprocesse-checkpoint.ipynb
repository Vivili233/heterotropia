# 缺失值

```python
from sklearn.impute import SimpleImputer
#使用众数填补斜视年龄
age=df.loc[:,"斜视年龄"].values.reshape(-1,1)
imp_mode=SimpleImputer(strategy="most_frequent")
df.loc[:,"斜视年龄"]=imp_mode.fit_transform(age)

df.info()
```

```python
age[:20]
```

```python
#使用0来填补有无同时视
from sklearn.impute import SimpleImputer
TSS=df.loc[:,"同时视有值"].values.reshape(-1,1)
TSS_no=df.loc[:,"同时视无值"].values.reshape(-1,1)
imp_0=SimpleImputer(strategy="constant",fill_value=0) #用0填补
df.loc[:,'同时视有值']=imp_0.fit_transform(TSS)
df.loc[:,'同时视无值']=imp_0.fit_transform(TSS_no)
df.head()
```

```python
#用中位数填补视近视远的缺失值
SJ=df.loc[:,"视近（直角）"].values.reshape(-1,1)
SY=df.loc[:,"视远（直角）"].values.reshape(-1,1)
imp_median=SimpleImputer(strategy="median")
df.loc[:,"视近（直角）"]=imp_median.fit_transform(SJ)
df.loc[:,"视远（直角）"]=imp_median.fit_transform(SY)
df.head()
```

# 处理分类型特征：编码与哑变量
将文字数据转换为数值数据

preprocessing.LabelEncoder:标签专用，能够将分类转换为分类数值

preprocessing.OrdinalEncoder:特征专用，能够将分类特征转换为分类数值（有序变量）

preprocessing.OneHotEncoder:读热编码，创建哑变量（名义变量）

```python
#将字符串类型转换成数值型，0表示无手术史，1表示有。
df["手术史"] = df["手术史"].map({"无":0,"有":1})
df.head()
```

```python
df.iloc[:,24]
```

```python
#肌肉方式转换为数值型，0表示无，1为后退，2为缩短
df["方式（右外直肌）"] = df["方式（右外直肌）"].map({"无":0,"后退":1,"缩短":2})
df["方式（左外直肌）"] = df["方式（左外直肌）"].map({"无":0,"后退":1,"缩短":2})
df["方式（右内直肌）"] = df["方式（右内直肌）"].map({"无":0,"后退":1,"缩短":2})
df["方式（左内直肌）"] = df["方式（左内直肌）"].map({"无":0,"后退":1,"缩短":2})
df["方式（右上直肌）"] = df["方式（右上直肌）"].map({"无":0,"后退":1,"缩短":2})
df["方式（左上直肌）"] = df["方式（左上直肌）"].map({"无":0,"后退":1,"缩短":2})
df["方式（右下直肌）"] = df["方式（右下直肌）"].map({"无":0,"后退":1,"缩短":2})
df["方式（左下直肌）"] = df["方式（左下直肌）"].map({"无":0,"后退":1,"缩短":2})
df["方式（右上斜肌）"] = df["方式（右上斜肌）"].map({"无":0,"离断":1,"转位":2})
df["方式（左上斜肌）"] = df["方式（左上斜肌）"].map({"无":0,"离断":1,"转位":2})
df["方式（右下斜肌）"] = df["方式（右下斜肌）"].map({"无":0,"离断":1,"转位":2})
df["方式（左下斜肌）"] = df["方式（左下斜肌）"].map({"无":0,"离断":1,"转位":2})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
zd=df.iloc[:,23]         #要输入的是标签，不是特征矩阵，所以允许一维
le=LabelEncoder()         #实例化
le=le.fit(zd)              #导入数据
le.classes_              #属性classes查看标签中究竟有多少类别
label=le.transform(zd)     #transform接口调取结果
```

```python
label
```

```python
df.iloc[:,23]=label
df.iloc[:,23]
```

```python
x=df.iloc[:,:24]
y=df.iloc[:,24]
from sklearn.model_selection import train_test_split
#x为数据集的feature熟悉，y为label.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
x_train.shape
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
predict_y = knn.predict(x_test)
# 评分
print("mean_squared_error:", mean_squared_error(y_test, predict_y))
print('accuracy_score：', accuracy_score(y_test, predict_y))
```

# 分类器集成

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import cross_val_score
```

```python
params = [1, 10, 15, 20, 25, 30, 40]
test_scores = []
for param in params:
    clf = BaggingRegressor(n_estimators=param, base_estimator=knn)
    test_score = np.sqrt(-cross_val_score(clf, x_train, y_train, cv=10, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(params, test_scores)
plt.title("n_estimator vs CV Error");
```
